```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
%matplotlib notebook
```

```python
X,y=make_regression(n_samples=1000, n_features=2,n_informative=1,noise=10,random_state=1)
print(X.shape,y.shape)
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)
plt.subplot(1,2,2)
plt.scatter(X[:,1],y)
```
